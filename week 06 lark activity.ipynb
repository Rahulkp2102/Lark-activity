```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel('Dataset_w6.xlsx')
```

# visualization

```python
plt.figure(figsize=(10,8))
sns.barplot(x='country',y='Population in thousands (2017)',data=data.sort_values('Population in thousands (2017)',ascending=False).head(10))
plt.title('countries vs population',fontsize=20)
plt.xlabel('countries')
plt.ylabel('population')
plt.xticks(rotation=90)
plt.show()
```

```python

plt.figure()
sns.barplot(x='country',y='Agricultural production index ',data=data.sort_values('Agricultural production index ').head(10),color='blue') 
plt.title('Top 10 countries VS Agricultural production',fontweight='bold',fontsize=18)
plt.xlabel('Country')
plt.ylabel('Agricultural production index')
plt.xticks(rotation=90)
plt.show()


```

```python
plt.figure(figsize=(8,4))
plt.scatter(data['Region'],data['International trade: Imports (million US$)'],s=10,c='green')
plt.title('Plot of Region VS International Trade',fontweight='bold',fontsize=18)
plt.xticks(rotation=90)
plt.xlabel('Region')
plt.ylabel('International Trade')
```

# no.of countries with respect to region

```python
data.groupby('Region')[['country']].count()
```

# region having highest

```python
#GDP per capita (current US$)
```

```python
gdp=data.groupby('Region')[['GDP per capita (current US$)']].mean()
gdp.sort_values('GDP per capita (current US$)',ascending=False)
```

```python
# International trade: Imports (million US$)
```

```python
iti=data.groupby('Region')[['International trade: Imports (million US$)']].mean()
iti.sort_values('International trade: Imports (million US$)',ascending=False)
```

# region more consistant with respect to GDP per capita (current US$)

```python
con_gdp=data.groupby('Region')[['GDP per capita (current US$)']].std()
con_gdp.sort_values('GDP per capita (current US$)',ascending=True)
```
